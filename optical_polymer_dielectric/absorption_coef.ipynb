```python
import matplotlib.pyplot as plt
import matplotlib.axes as ax
import seaborn
import numpy as np
import pandas as pd
from IPython import display
```

```python
z15y0 = np.loadtxt('z1.5_y0/zpol/absorp_coef.out')
z35y0 = np.loadtxt('z3.5_y0/zpol/absorp_coef.out')
z65y0 = np.loadtxt('z6.5_y0/zpol/absorp_coef.out')


fig, ax = plt.subplots(1,3,figsize=(10,5), layout='tight', sharex='all', sharey='all')

plt.suptitle('Absorption coefficients Across Z-distances 1.5-6.5 at Y=0')

plt.subplot(1,3,1)
plt.title('z=1.5')
plt.ylabel('Absorption Coefficients (cm^(-1))')
plt.plot(z15y0[:,0],z15y0[:,1])

plt.subplot(1,3,2)
plt.title('z=3.5')
plt.plot(z35y0[:,0],z35y0[:,1])

plt.subplot(1,3,3)
plt.title('z=6.5')
plt.plot(z65y0[:,0],z65y0[:,1])

plt.show()
```
